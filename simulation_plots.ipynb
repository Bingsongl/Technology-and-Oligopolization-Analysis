```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

# Simulated Results

Simulated results for the expected distributions of each hypothesized results for the paper

## Figure 1

```python
num_points = 1000
xmin = 1  # Minimum value
alpha = 2.5  # Exponent

x = np.random.poisson(10000,1000)

# Plotting
plt.figure(figsize=(4,4))
plt.hist(x, bins=50, density=True, alpha=0.7, color='b')

# Plotting the theoretical power law curve
plt.title('Ideal Poisson Distribution (Non-Oligopoly)')
plt.xlabel('Out-Degree')
plt.ylabel('Number of Countries')
plt.grid(True, which="both")

plt.xticks([])
plt.yticks([])

plt.show()
```

```python
num_points = 1000
xmin = 1  # Minimum value
alpha = 10  # Exponent

x = np.random.pareto(alpha, num_points) + xmin

# Plotting
plt.figure(figsize=(4,4))
plt.hist(x, bins=50, density=True, alpha=0.7, color='b')

# Plotting the theoretical power law curve
plt.title('Ideal Power Law Distribution (Oligopoly)')
plt.xlabel('Out-Degree')
plt.ylabel('Number of Countries')
plt.grid(True, which="both")

plt.xticks([])
plt.yticks([])

plt.show()
```
